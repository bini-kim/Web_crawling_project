## 종합순위 출력

### 베스트셀러 목록 url사용하여 종합순위 목록의 주소 리스트 출력하기!

```python
#목록 url
url = 'http://book.interpark.com/display/collectlist.do?_method=bestsellerHourNew&bookblockname=b_gnb&booklinkname=%BA%A3%BD%BA%C6%AE%C1%B8&bid1=w_bgnb&bid2=LiveRanking&bid3=main&bid4=001'

#url에 get요청하기
resp = requests.get(url)

#html
html = BeautifulSoup(resp.text, 'html.parser')

#html 태그 사용
#html.select('div.listItem.singleType')[1]

#각 목록의 주소 리스트 생성
anchors = []
for href in html.select('div.listItem.singleType'):
    anchors.append(href.find('a').get('href'))
anchors
```

## 주소 내 정보 가져오기

```python
#(첫번째 주소) 테스트 하기
a = 'https://book.interpark.com/product/BookDisplay.do?_method=detail&sc.shopNo=0000400000&sc.prdNo=354750490&sc.saNo=003003001&bid1=Best_zone&bid2=LiveRanking&bid3=PRD&bid4=001'
resp = requests.get(a)
soup = BeautifulSoup(resp.text, 'html.parser')

#제목
print(soup.select('h2')[0].text)

#가격
print("정가 : "+ soup.select('p.fixedPrice')[0].text)

#정보
print(soup.select('ul.bInfo_txt')[0].text)
```

## 종합순위 내 상위 15위 목록 출력하기

```python
#베스트셀러 종합순위 출력

# 딕셔너리 담을 리스트 생성
best_seller = []

# anchors url 사용
for anchor in anchors:
    resp = requests.get('http://book.interpark.com'+anchor)
    soup = BeautifulSoup(resp.text, 'html.parser')

    #index사용해서 순서 출력
    print(f'----------------{anchors.index(anchor)+1}위------------------')

    #딕셔너리 생성
    dic_info = {}

    dic_info['도서명'] = soup.select('h2')[0].text
    print("도서명 : " + dic_info['도서명'])
    dic_info['가격'] = soup.select('p.fixedPrice')[0].text
    print("정가 : " + dic_info['가격'])
    dic_info['저자'] = soup.select('.bInfo_txt li a')[0].text
    print("저자 : "+dic_info['저자'])
    dic_info['출판사'] = soup.select('.bInfo_txt li')[-4].text
    print(dic_info['출판사'])
    dic_info['발행일'] = soup.select('.bInfo_txt li')[-3].text
    print(dic_info['발행일'])
    dic_info['페이지'] = soup.select('.bInfo_txt li')[-2].text
    print(dic_info['페이지'])
    dic_info['ISBN'] = soup.select('.bInfo_txt li')[-1].text
    print(dic_info['ISBN'])
    dic_info['판매지수'] = soup.select('div.indexBox span')[0].text
    print("판매지수 : "+ dic_info['판매지수'])
    dic_info['분야']='종합'
    print("분야 : "+dic_info['분야'])

    #dic_info에 best_seller 리스트 append
    best_seller.append(dic_info)

    #타이머
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(3)
    time.sleep(a)
print('---------------- result ------------------')
print(best_seller)
```
